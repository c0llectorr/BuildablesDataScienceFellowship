## **Task Overview**

In this task, I am going to manipulate the titanic dataset by: 
1. Dropping unnecessary columns.
2. Handling Missing Values by either replacing it via mean, median and mode imputation or removing from the data-frame.
3. Dropping unnecessary rows.

```python
import pandas as pd
```

Reading Titanic CSV File.

```python
df = pd.read_csv('titanic/train.csv')
```

#### Getting insight of the dataset by viewing the head.

```python
df.head()
```

#### **Data Cleaning by Dropping Un-necessary columns.**

```python
df.drop(columns=["PassengerId", "SibSp", "Parch"], inplace=True)
```

```python
df.head()
```

#### **Replacing Not a Number (NaN)s Values with Unknown which is more suitable and can help in categorizing values.**

```python
df['Cabin'] = df['Cabin'].fillna('Unknown')
df.head()
```

#### Again replaced ambiguous or misleading enteries about ports with their actual name
#### for better understanding of the data and useful categorization.

```python
df.replace(['C', 'Q', 'S'], ['Cherbourg', 'Queenstown', 'Southampton'], inplace=True)
df.head()
```

#### Replacing NaN values with median imputation Value to get more realistic central tendency
#### Handles skewed data and avoid distortion from extreme ages.

```python
#Checking if the file contains any NaN Value.
df['Age'].isna()
#Replacing the NaN Values with median imputation.
median_imputation = df['Age'].median()
df['Age'] = df['Age'].fillna(median_imputation)
```
