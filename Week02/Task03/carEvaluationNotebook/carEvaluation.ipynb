## **Task Overview**

**In this task I have been given with a car evaluation dataset.**

**I have to pre-process the data in the following formats and analyse the results:**

**1. OneHotEncoding**

**2. LabelEncoding** 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('carData.csv')
```

**After reading the data from the file, let's just check what features/columns does the dataset contain.**

```python
df.head()
```

**As we can see, the dataFrame doesn't contain any columns so lets assign the dataFrame column names for better understanding.**

```python
#Creating a list of columns.
cols = ['Buying', 'Maintenance', 'Doors', 'Persons', 'Luggage Boot', 'Safety', 'Class']
df.columns = cols
df.head()
```

**Converting the dataFrame into OneHotEncoding Values via pandas get_dummies() function.**

```python
df_oneHotEncoded = pd.get_dummies(df, dtype=int)
df_oneHotEncoded.head()
```

**Label Encoding of the original data frame via sklearn library.**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lenc = LabelEncoder()
lenc_df = df.copy()
for col in cols:
    lenc_df[col] = lenc.fit_transform(lenc_df[col])
    
lenc_df.head()
```

**As we can see Label Encoder has less dimensions than OneHotEncoder.**

**In this car evaluation dataset we must use LabelEncoder instead of OHE because the features of the dataframe are ordinal.**
